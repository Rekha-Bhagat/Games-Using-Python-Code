```python
def user_choice():
    
    # variables
    
    #Initial    
    choice = 'wrong'
    acceptable_range=range(0,10)
    within_range=False 
    
    # Two Conditions to check.
    # Digit Or Within_range==False
    
    while choice.isdigit() ==False or within_range==False:
        choice=input("Please Enter a number (0-10): ")  #9
        
        # Digit Check
        if choice.isdigit()==False:
            print("Sorry Wrong input has been given ..\n Please enetr a digit between 0 - 10")
            
        # Range Check
        if choice.isdigit()==True:
            if int(choice) in acceptable_range:
                within_range=True
                print("Well done Entered Correct Choice....")
            else:
                within_range=False
                print("The number is not within the range.. \nPlease Enter a number within 0-10")
            
    return int(choice)
```
